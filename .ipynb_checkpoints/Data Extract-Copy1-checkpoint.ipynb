```python
df2 = df_pc[0][['County /','Unnamed: 4','Use of Desktop Computer/Laptop/Tablet','Unnamed: 5']]
```

```python
df_new = df2.copy()
```

```python
df_new.iloc[:1,-2:-1] = "ttl_percent Male percent Female"
```

```python
df_new.head()
```

```python
import re
```

```python
def space_separator(row,row_num):
    """
    row = is the rows in the dataframe df
    row_num = is the index of the row
    """
    #clearing  lists
    items = []
    sorted_items = []
    
    #splitting the  row
    items = re.split(' ',row)

    #removing extra spaces and storing in sorted items list
    sorted_items = [item for item in items if item != '']
    
    #setting the list items into columns in the data frame
    df_new.loc[row_num,'total_percent'] = sorted_items[0]
    df_new.loc[row_num,'Male'] = sorted_items[1]
    df_new.loc[row_num,'male_percent'] = sorted_items[2]
    df_new.loc[row_num,'Female'] = sorted_items[3]
    
    #printing out the sorted row
    print(sorted_items)
```

```python
#iterating through the rows in the dataframe
for row_num in range(len(df_new['Use of Desktop Computer/Laptop/Tablet'])):
    space_separator(df_new['Use of Desktop Computer/Laptop/Tablet'][row_num], row_num)
```

```python
df_new.head()
```

```python
df_final = df_new[1:8][['County /','Unnamed: 4','total_percent','Male','male_percent','Female','Unnamed: 5']]
```

```python
df_final.columns
```

```python
df_final = df_final.rename(columns={"County /": "county/subcounty", 
                         "Unnamed: 4": "total pop using pc's", 
                         "total_percent": "percent of total pop age3+", 
                         "Male": "males", 
                         "male_percent": "percent of males",
                         "Female": "females",
                         "Unnamed: 5": "percent of females"})
```
